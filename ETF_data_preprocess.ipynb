```python
from datetime import datetime
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['month'] = df['date'].dt.month
df['week'] =[pd.to_datetime(df.date[i]).weekday() for i in range(len(df.date))]
```

```python
df.head()
```

```python
td = [1, 3, 5, 10, 20, 40, 60, 120, 250]
```

```python
df.columns
```

```python
df_col = ['SPY_p', 'IWM_p', 'EEM_p', 'TLT_p', 'LQD_p', 'TIP_p', 'IYR_p',
       'GLD_p', 'FXE_p', 'SPY_v', 'IWM_v', 'EEM_v', 'TLT_v', 'LQD_v', 'TIP_v',
       'IYR_v', 'GLD_v', 'FXE_v']
```

```python
%%time

for k in df[df_col]:
    
    print(k," Processing...")

    for i in td:
        
        #print(i)

        cir_check = 1        

        a = k+"_"+str(i)            


        for j in range(i,len(df)):
            
            #print(j)

            if  cir_check == 1:

                df[a] = 0.0

                cir_check = 0

            else:

                pass


            if k.find('_v') == -1:

                df[a][j] = float((df[k][j] / df[k][j-i]) - 1)


            else:

                df[a][j] = df[k][j-i:j].mean()


```

```python
df_col_p = ['SPY_p',
            'IWM_p',
            'EEM_p',
            'TLT_p',
            'LQD_p',
            'TIP_p',
            'IYR_p',
            'GLD_p',
            'FXE_p']
```

```python
%%time

for i in df_final[df_col_p]:
    
    print(i," Processing...")
        
    for j in td:
        
        t_name = str(i)+"_"+str(j)
        y_name = str(i)[0:3]+"_"+str(j)+"_y"

        df_final[y_name] = 0.0

        for j in range(1,len(df_final)):
                        
            

            if df_final[t_name][j] > 0 :
                
                

                df_final[y_name][j-1] = "1"

            else:
                                

                df_final[y_name][j-1] = "0"

            #print(df_final[i][j])

            #print(str(i)[0:3]+"_y")
```
